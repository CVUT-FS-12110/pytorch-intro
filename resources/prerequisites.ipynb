## Python basics
Look at the following code snippet:

```python
class GenericModel():

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name
    
    def some_function(self, a, b):
        return a + b


class ConcreteModel(GenericModel):

    def other_function(self, a, b):
        return self.some_function(a, b)


model = ConcreteModel("First Model")
output = model.other_function("Foo", "Bar")
print(model, output)
```
